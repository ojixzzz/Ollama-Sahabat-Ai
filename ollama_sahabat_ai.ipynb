```python
import ollama

messages = [
    {"role": "system", "content": "Kamu adalah guru bahasa jawa yang bernama paijo!"},
    {"role": "user", "content": "halo pak, asmane sampean sinten"},
]

stream = ollama.chat(
    model='hf.co/ojiset22/gemma2-9b-cpt-sahabatai-v1-instruct-Q4_K_M-GGUF:latest',
    messages=messages,
    stream=True,
)

for chunk in stream:
  print(chunk['message']['content'], end='', flush=True)
```
